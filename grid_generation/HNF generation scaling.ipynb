```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
```

```python
columns = ["N cells","Removing dups (sec)","Mink reduction (sec)","Just HNF generation (sec)"]
index = ["10","20","30","40","50","100","1000","2000","3000","5000","10000","15000"]
```

```python
data = np.array([np.zeros(12)]*4).T
# N cells
data[0][0] = 10
data[1][0] = 20
data[2][0] = 30
data[3][0] = 40
data[4][0] = 50
data[5][0] = 100
data[6][0] = 1000
data[7][0] = 2000
data[8][0] = 3000
data[9][0] = 5000
data[10][0] = 10000
data[11][0] = 15000

#Mink and dup reduction times
data[0][1] = 0.1679
data[1][1] = 3.0801
data[2][1] = 17.6323
data[3][1] = 49.1725
data[4][1] = 60.9685

#Mink only times
data[0][2] = 0.007522
data[1][2] = 0.0163769
data[2][2] = 0.041191
data[3][2] = 0.062014
data[4][2] = 0.074299
data[5][2] = 0.34176
data[6][2] = 40.71971

#Only finding HNFs
data[0][3] = 0.00319
data[1][3] = 0.002387
data[2][3] = 0.00257199
data[3][3] = 0.00319699
data[4][3] = 0.00280699
data[5][3] = 0.004926
data[6][3] = 0.23551799
data[7][3] = 0.96137499
data[8][3] = 2.96218824
data[9][3] = 5.80610895
data[10][3] = 15.99445927
data[11][3] = 81.852699
```

```python
df = pd.DataFrame(data, index=index,columns=columns)
```

```python
df
```

```python
df.iloc[0:4].plot(x="N cells",y="Removing dups (sec)",
                  title="HNF generation timing with removal of duplicates and minkowski reduction.")
plt.savefig('Removing_duplicates.pdf')
plt.show()
```

```python
df.iloc[0:6].plot(x="N cells",y="Mink reduction (sec)",
                  title="HNF generation timing with minkowski reduction.")
plt.savefig('Mink_reduction.pdf')
plt.show()
```

```python
df.plot(x="N cells",y="Just HNF generation (sec)",
                  title="HNF generation timing only.")
plt.savefig('HNF_only.pdf')
plt.show()
```
